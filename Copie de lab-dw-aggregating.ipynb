6.  Display a new DataFrame that contains the number of policies sold by month, by state, for the top 3 states with the highest number of policies sold.

*Hint:*
- *To accomplish this, you will first need to group the data by state and month, then count the number of policies sold for each group. Afterwards, you will need to sort the data by the count of policies sold in descending order.*
- *Next, you will select the top 3 states with the highest number of policies sold.*
- *Finally, you will create a new DataFrame that contains the number of policies sold by month for each of the top 3 states.*

```python
#Display a new DataFrame that contains the number of policies sold by month, by state, for the top 3 states with the highest number of policies sold.

# Group by state and sum the number of policies to find the total for each state
state_policy_counts = df.groupby("state")["number_of_policies"].sum()

# Sort the states by the total number of policies in descending order and select the top 3
top_3_states = state_policy_counts.sort_values(ascending=False).head(3).index

# Filter the original DataFrame to include only the top 3 states
df_top_states = df[df["state"].isin(top_3_states)]

# Group the filtered data by state and month, sum the policies, and unstack to get the desired format
df_policies_by_month_state = df_top_states.groupby(["state", df_top_states["effective_to_date"].str.split('/').str[0]])["number_of_policies"].sum().unstack()

df_policies_by_month_state
```
